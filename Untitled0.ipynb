```python
n = 100
import numpy as np
import math
y = np.random.randint(0,2,(1,n))
ycap = np.random.uniform(0,1,size = (1,n))
j = math.log(2)
one = np.ones((1,n))
oneycap = one - ycap
logycap = np.log(ycap)/j
logoneycap = np.log(oneycap)/j
Op = -(1/n)*(y*logycap+oneycap*logoneycap)
O = 0
for i in range(0,n):
  O = O + Op[0,i]
print(Op)
print(y)
print(ycap)
print(O)
```
